# 최종 기여도그룹 부여 ( 1.라벨 )

```python
labels_ = np.array([])

for col in tdKMeans.cluster_info.columns:
#     max_group = tdKMeans.cluster_info[col].value_counts().idxmax()
    max_group = tdKMeans.cluster_info[col].mean().round().astype("int")
    
    labels_ = np.append(labels_,max_group)
    
labels_.astype("int")
```

# 최종 기여도그룹 번호 부여

```python
_unique_labels = np.unique(tdKMeans.labels_)
groups_ = np.zeros(len(labels_)) - 1

for idx, label in enumerate(_unique_labels):
    groups_[labels_ == label] = idx
    
groups_
```

# 최종 클러스터링 패턴 선정

```python
# 0번 그룹 친구들
group_df = tdKMeans.cluster_info.iloc[:,tdKMeans.groups_ == 0]
```

```python
_unique_groups = np.unique(groups_)
```

```python
clusters_ = np.array([])

for _group in _unique_groups:
    group_df = tdKMeans.cluster_info.iloc[:,tdKMeans.groups_ == _group]

    for _round in range(len(tdKMeans.kmeans_)):
        max_cluster = group_df.iloc[_round].value_counts().idxmax()

        clusters_ = np.append(
            clusters_,
            tdKMeans.kmeans_[_round].clusters_[max_cluster]
        )

print(clusters_.shape)
clusters_[:5]
```

```python
clusters_ = clusters_.reshape(len(_unique_groups), -1)
clusters_
```
