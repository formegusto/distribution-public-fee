```python
from _src import SimpleDistributor, GroupDistributor
```

```python
import pandas as pd

file_path = "ami_datas/apt_780_201809.csv"
meter_month = pd.read_csv(file_path)
```

```python
meter_month.columns = ['name', 'usage (kWh)']
meter_month.head()
```

```python
sim_dist = SimpleDistributor(file_path=file_path)

sim_dist.month = 9
sim_dist.meter_month = meter_month
```

```python
sim_dist.set_cont()
```

```python
households_kwh = meter_month['usage (kWh)'].sum()
public_kwh = households_kwh * 0.25
apt = households_kwh + public_kwh

sim_dist.set_calc(kwh=apt)
```

```python
dist_df = sim_dist.distribute_table()
```

```python
dist_df.sort_values('기여도').head()
```
